## **Naming conventions**

In Python, it is convention to use **snake case** for variable and function names. Snake case is generally characterized by all lowercase letters, separated by underscores, e.g., `snake_case_is_cool`.

All object names in Python are case sensitive, so be careful to use the correct names and cases when using the language.

## **Typecasting**

**Typecasting**, or type conversion, is the process of converting one data type to another. Typecasting is useful if you need all your values to have the same data type.

To typecast one data type to another, use the type constructor function for the data type you want. The type constructor function is the name of the data type followed by parentheses.

```python
# create an integer
num_a = 11

# typecast the integer to a float with the float() constructor function
num_a = float(num_a)

# print the new value and its data type
print(num_a)
print(type(num_a))
```

Keep in mind that there is possible data loss with typecasting. For example, the process of converting a float to an int loses decimal precision.

```python
num_b = 2.75
num_b = int(num_b)

print(num_b)
print(type(num_b))
```

Additionally, typecasting does not always work. A string containing alphabetical characters, for example, cannot be converted to a numeric type because they are represented differently in memory.

```python
some_string = "77abc"

some_string = int(some_string)
```

## **Overloaded operators**

In Python, some of these operators are **overloaded**. This means that the same operator can be used on different data types.

Let's try a couple of our arithmetic operators on the string data type!

The subtraction `-` and division `/` operators do not work with the string data type.

```python
# Try to print "2"-"3"
print("2"-"3")
```

```python
# Try to print "2"/"3"
print("2" / "3")
```

We can't use subtraction or division on strings.

The addition `+` operator, however, does work on strings!

```python
# Try to print "2"+"3"
print("2"+"3")
```

### **String concatenation**

While the `+` operator did not result in an error in the previous example, notice that we did not add 2 and 3 together. Instead, we performed a **string concatenation**. String concatenation is the process of combining individual strings together to form one long string.

In the above example, we concatenated the individual strings "2" and "3" to get the combined result "23" using the `+` operator.

In Python, the `+` operator is overloaded - it can be use for both numeric addition and string concatentation.
